```python
import numpy as np
import matplotlib.pyplot as plt

days = [1,2,3,4,5,6,7]

np.random.seed(40)

temperature = np.random.rand(7)

temp_calc = temperature * 100

temp_calc

plt.plot(days, temp_calc, marker='o')
plt.xlabel('days')
plt.ylabel('temperature')
plt.title('Daily Temperature Report')
plt.legend('days')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

products = ['Product A', 'Product B', 'Product C', 'Product D']

np.random.seed(40)
sales = np.random.rand(4)

sales_final = sales * 100

plt.bar(products, sales, color='r')
plt.xlabel('products')
plt.ylabel('sales')
plt.title('Sales by Product Report')
plt.grid(True)
plt.show()
```
