```python
from selenium import webdriver
from bs4 import BeautifulSoup
from selenium.webdriver.chrome.service import Service
import csv

driver = webdriver.Chrome()

#create the url
page_url = "https://www150.statcan.gc.ca/t1/tbl1/en/tv.action?pid=1810000401"

#fetch the page
driver.get(page_url)

#wait for the table to load (you might need to adjust the wait time)
driver.implicitly_wait(10)

#get the page source AFTER JavaScript execution
page_source = driver.page_source

#Parse the HTML content
soup = BeautifulSoup(page_source,'html.parser')

#Find the HTML/table element
table = soup.find('table', class_='pub-table')

if table:
    #extract the column headers
    headers = [th.get_text().strip() for th in table.find_all('th')]

    #extract the row data
    rows = []
    for tr in table.find_all('tr'):
        row = [td.get_text().strip() for td in table.find_all('td')]
        if row: #check if row is NOT empty
            rows.append(row)

    #print column headers
    print("Column Headers")
    print(headers)

    #print rows
    print("\n Row Data:")
    for row in rows:
        print(row)

    headers_tuple = (
        'All-items',
        'Food',
        'Shelter',
        'Household operations, furnishings and equipment',
        'Clothing and footwear',
        'Transportation',
        'Gasoline',
        'Health and personal care',
        'Recreation, education and reading',
        'Alcoholic beverages, tobacco products and recreational cannabis',
        'All-items excluding food and energy',
        'All-items excluding energy',
        'Energy',
        'Goods',
        'Services'
    )

    #write the data to a csv file
    with open('table_data.csv','w', newline='') as csvfile:
        writer = csv.writer(csvfile)

        #insert the new column as the first column in each row
        for i, row in enumerate(rows):
            rows[i]=[headers_tuple[i]] + rows

        #write headers
        writer.writerow(headers[2:8])

        #write rows
        writer.writerows(rows)

    print("Data saved to table_data.csv successfully")

else:
    print("Table not found")

#Close the WebDriver
driver.quit()
        
```
